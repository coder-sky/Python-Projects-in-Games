```python
import random
```

```python
def hangman_lifes():
    lives = ['''
  ________
   |     \|
   O      |
  /|\     |
  / \     |
          |
  ==LOST===
 ''' , '''
  ________
   |     \|
   O      |
  /|\     |
    \     |
          |
  =========
  ''' , '''
  ________
   |     \|
   O      |
  /|\     |
          |
          |
  =========
  ''','''
  ________
   |     \|
   O      |
   |\     |
          |
          |
  =========
  ''','''
  ________
   |     \|
   O      |
   |      |
          |
          |
  =========
   ''','''
  ________
   |     \|
   O      |
          |
          |
  =========
  ''',''' 
  ________
   |     \|
          |
          |
          |
          |
  =========
   ''']
    game_name = '''
    
 _                                             
| |                                            
| |___    __ _      __ __    __ _     _____      __       ___  
|  _  |  / _  |    |  _  \  / _  |  /_   _ \   /  _\    |  _ \ 
| | | | | (_| |_   | | | | | (_| |  | | | | | | (_| |_  | | | |
|_| |_|  \______|  |_| |_|  \___ |  |_| | |_|  \______| |_| |_|
                             __/ |                      
                            |___/

'''
    return lives,game_name
```

```python
def hangman_guess():
    guess_list = ['Almond', 'Banana', 'Blewits', 'Brazil', 'Carrot', 'Casaba', 'Cashew', 'Celery', 'Cherry', 'Chives', 'Chocho', 'Citron', 'Citrus', 'Cob nut', 'Cooker', 'Daikon', 'Damson', 'Endive', 'Fennel', 'Garlic', 'Girkin', 'Greens', 'Kiwano', 'Lentil', 'Lichee', 'Lychee', 'Marrow', 'Medlar', 'Nettle', 'Orange', 'Oyster', 'Papaya', 'Pawpaw', 'Peanut', 'Pepper', 'Pignut', 'Pippin', 'Potato', 'Quince', 'Radish', 'Raisin', 'Rocket', 'Russet', 'Squash', 'Tomato', 'Turnip', 'Wakame', 'Walnut']
    return guess_list
```

```python
def hangman_main():
    g_list=hangman_guess()  #object of guessing list
    g_word=random.choice(g_list).lower()   #obj of guessing word
    word_letter=len(g_word)
    
    game_over=False
    tries=6
    
    lifes,name=hangman_lifes()
    print(name)  #hangman logo printing
    print('Game Starts'.center(100,'-'))
    print("RULES:\n1.You Need to guess right word\n2.See the given hint and answer\n3.Enter the letters sequentially\n4.Enter only one valid letter at a time")
    print('-'*100)
    
    hint=''  #Adding hint
    for i in range(word_letter):
        if i==1:
            hint+=g_word[i]
        elif i==4:
            hint+=g_word[i]
        else:
            hint+='_'
    print("Complete the whole by using following Hint\nHINT:",hint)
    #print(g_word)
    
    result=[]
    for i in range(word_letter):
        result += "_"
    #print(result)
    count=0
    
    while not(game_over):
        word=''
        user_guess=input("Guess a letter:").lower()
        while len(user_guess)!=1:
            user_guess=input("Guess a single letter at a time:").lower()
        temp=True
        for pos in range(len(g_word)):
            letter=g_word[pos]
            
            if letter==user_guess:
                result[count]=letter
                count+=1
                break
        if g_word[count-1]==user_guess: #checking right position
            temp=False
            for i in result:
                word+=i
            print("WORD=",word)
            
        if temp:
            tries -= 1
            print(f"You guessed {user_guess}, this letter is not in the word, You lose a try.\nLifes Left:{tries}")  
            
            if tries==0:
                print("You Lose!Game Over\nAnswer=The Right Word is:",g_word)
                game_over=True
        elif '_' not in result:
            game_over = True
            print("You are a winner, Congratulations.")
        print(lifes[tries])
```

```python
hangman_main()
```
